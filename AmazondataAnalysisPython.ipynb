```python
#Reading the dataset
df=pd.read_csv('amazonDA.csv')
```

```python
#Displaying the first 15 rows 
df.head(10)
```

```python
#Checking for missing values 
df.isnull().sum()
```

```python
#Remove the commas  from rating_count column
df['rating_count'] =df['rating_count'].astype(str).str.replace(',','')

# Convert rating_count to numeric 
df['rating_count']= pd.to_numeric(df['rating_count'],errors ='coerce')

#Determine the mean of the rating_count
mean_rating_count = df['rating_count'].mean()

#Replace the missing values with the mean of the rating_count column 
#Why: The method ensures that the overall distribution of amazon data is insignificantly affected by the missing values
df['rating_count'].fillna(mean_rating_count)
df.head(5)
```

2. Data Cleaning 

```python
#Remove the commas from the columns
df['discounted_price'] = df['discounted_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['actual_price'] = df['actual_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
```

```python
#Convert the columns to numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'])
df['actual_price'] = pd.to_numeric(df['actual_price'])
```

```python
#Extract relevant product categories from category column
df['main_category'] =df['category'].str.split(',').str[0]
df['sub_category'] =df['category'].str.split(',').str[1]
```

3. Data Transformation

```python
#Create a new column for discount_amount (actual_price -
#discounted_price).
df['New_discoounted_price'] = df['actual_price'] - df['discounted_price']
```

```python
#Convert discount_percentage into a numeric value
#Remove the '%' sign and convert the discoun_% into numeric
df['discount_percentage'] =df['discount_percentage'].str.replace('%','').astype(float)
df.head(5)
```

```python
#Extract the year of reviews if a timestamp is present.
#check if review_stamp exist
if'review_stamp' in df.columns:
#Extract the year from the review_stamp
   df['review_year'] = df['review_stamp'].dt.year
else:
   print("The 'review_stamp' column does not exist in the data frame")
```
